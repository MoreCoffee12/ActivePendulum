### Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

### Define command parameters

```python
dAmp = 90.0;
```

### The conventional approach

```python
t = np.arange(0.0, 256.0, 0.01)
```

```python
dCommand = dAmp*np.sin(2*np.pi*t/256.0)
dCommand = dCommand+90
```

### Plo the data out

```python
fig = plt.figure(1)
ax1 = fig.add_subplot(111)
ax1.plot(t, dCommand)
ax1.grid(True)
ax1.set_ylim((0, 180))
ax1.set_ylabel('Command value, counts')
ax1.set_title('Command versus Time')
```

### Going ballistic now

```python
fsync = 100;
a1 =  2 * np.cos( 2 * np.pi / fsync );
x1 = np.zeros(fsync);
a2 =  2 * np.cos( 4 * np.pi / fsync );
x2 = np.zeros(fsync);
xp = np.ones(len(x1));
```

### Calculate initial conditions

```python
x1[0] = dAmp*np.cos(0);
x1[1] = dAmp*a1/2.0;

x2[0] = np.cos(0);
x2[1] = a2/2.0;

xp[0] = xp[0] - x1[0] - x2[0];
xp[1] = xp[1] - x1[1] - x2[1];
```

```python
for idx in np.arange(2,fsync):
    
    x1[idx] = a1 * x1[idx-1] - x1[idx-2];
    x2[idx] = a2 * x2[idx-1] - x2[idx-2];
    xp[idx] = xp[idx] - x1[idx] - x2[idx];
```

```python
fig = plt.figure(2)
ax1 = fig.add_subplot(111)
ax1.plot(x1+90)
ax1.grid(True)
ax1.set_ylim((-1, 180))
ax1.set_ylabel('Command value, counts')
ax1.set_title('Command versus Time')
```
